# Quick example to fit a baseline

Signal is not standardized in rampy contrary to Spectra.jl, so behavior of splines can differ for the same smoothing coefficients.

```python
%pylab inline
#import sys
sys.path.append("../")
import numpy as np
import scipy
from matplotlib import pyplot as plt
#import gcvspline
import rampy
from sklearn import preprocessing
```

```python
nb_points  =500
x = np.linspace(50, 500, nb_points)

# gaussian peaks
p1 = 20.0 * np.exp(-np.log(2) * ((x-150.0)/15.0)**2)
p2 = 100.0 * np.exp(-np.log(2) * ((x-250.0)/5.0)**2)
p3 = 50.0 * np.exp(-np.log(2) * ((x-450.0)/1.0)**2)
p4 = 20.0 * np.exp(-np.log(2) * ((x-350.0)/30.0)**2)
p5 = 30.0 * np.exp(-np.log(2) * ((x-460.0)/5.0)**2)

# background: a large gaussian + linear 
bkg = 60.0 * np.exp(-np.log(2) * ((x-250.0)/200.0)**2) + 0.1*x

#noise
noise = 2.0 * np.random.normal(size=nb_points)

#observation
y = p1 + p2 + p3 + p4 + p5 + noise +bkg

plot(x,y,"r-")
```

# Different baseline examples

```python
# need to define some fitting regions for the spline
roi = np.array([[0,100],[200,220],[280, 290],[420,430],[480,500]])

# calculating the baselines
ycalc_gcvspl, base_gcvspl = rampy.baseline(x,y,roi,'gcvspline',s=100.0 )
ycalc_uni, base_uni = rampy.baseline(x,y,roi,'unispline',s=1e5)
ycalc_als, base_als = rampy.baseline(x,y,roi,'als',lam=10**6,p=0.05)
ycalc_arpls, base_arpsl = rampy.baseline(x,y,roi,'arPLS',lam=10**6,ratio=0.001)

# doing the figure
figure(figsize=(10,10))
plot(x,y,"k-",label="Raw")
plot(x,base_uni,"r-",label="gcvspline baseline")

plot(x,base_als,"-",color="purple",label="als baseline")
plot(x,base_arpsl,"-.",color="cyan",label="arPLS baseline")

xlabel("X")
ylabel("Y")
```

```python
nb_points  =500
x2 = np.linspace(100, 200, nb_points)
y_true = 10.0 * np.exp(-np.log(2) * ((x2-150.0)/15.0)**2) 
noise = 0.1 * np.random.normal(size=nb_points)
        
y_true_noise = y_true + noise
        
y2 = y_true_noise + 0.1*x2

roi = np.array([[100,110],[190,200]])
        
ycalc1, base = rampy.baseline(x2,y2,roi,'poly',polynomial_order=1)
ycalc2, base = rampy.baseline(x2,y2,roi,'unispline',s=0.0000001)
ycalc3, base = rampy.baseline(x2,y2,roi,'gcvspline',s=2.0)
ycalc4, base = rampy.baseline(x2,y2,roi,'als')
ycalc5, base = rampy.baseline(x2,y2,roi,'arPLS')

plt.plot(x2,y2,'k-')
plt.plot(x2,ycalc1)
```
